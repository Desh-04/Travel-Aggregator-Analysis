## **3. Answer to the questions:**

The data has been cleaned throughly so as to get accurate results. Lets now answer the following questions.

## 1)Find the number of distinct bookings,sessions,and searches from the given datasets.

```python
# Count the Distinct bookings
Distinct_bookings = booking['booking_id'].nunique()
print(f"Number of Distinct Booking: {Distinct_bookings}")

# Count the Distinct sessions
Distinct_sessions = session['session_id'].nunique()
print(f"Number of Distinct sessions: {Distinct_sessions}")

# Count the Distinct searches
Distinct_searches = session['search_id'].nunique()
print(f"Number of Distinct searches: {Distinct_searches}")
```

## 2)How many sessions have more than one booking?

```python
# Join bookings with sessions to find sessions with bookings
Mergedata = pd.merge(booking, session, on="booking_id", how="inner")

# Group by session_id and count the number of bookings in each session
Session_booking = Mergedata.groupby("session_id")["booking_id"].count()

# Filter sessions with more than one bookings
Sessions_with_multiple_bookings = Session_booking[Session_booking > 1] 

# Get the count of such sessions
Num_sessions_with_multiple_bookings = len(Sessions_with_multiple_bookings)

print("Number of sessions with more than one booking is:", Num_sessions_with_multiple_bookings)
```

## 3)Which days of the week have the highest number of bookings? Also, draw a piechart to show the distribution for all days of the week?

```python
# Extract the day of the week
booking['Day_of_week'] = booking['Booking_Date'].dt.day_name()

# Count bookings per day
booking_counts = booking['Day_of_week'].value_counts()

# Find the day with the highest bookings
Highest_booking_day = booking_counts.idxmax()
Highest_booking_count = booking_counts.max()

print(f"The highest number of bookings occur on: {Highest_booking_day} with {Highest_booking_count} bookings.")

# Ploting the pie chart
plt.figure(figsize=(8, 6))
plt.pie(booking_counts, labels=booking_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title("Bookings Distribution by Day of the Week")
plt.show()
```

## 4) For each of the service names, display the total number of bookings and the total Gross Booking Value in INR.

```python
# Group by service name(Booking_Platform) and calculate total bookings and total Gross Booking Value(GBV)
Booking_summary = booking.groupby('Booking_Platform').agg(Total_bookings=('booking_id', 'count'),Total_GBV=('INR_Amount', 'sum')).reset_index()

# Display the results
print(Booking_summary)
```

## 5) For customers who have more than 1 booking, which is the most booked route (from_city to to_city)?

```python
# Counting the number of bookings per customer
Booking_counts = booking['customer_id'].value_counts()

# Filtering customers with more than 1 booking
Multiple_bookings_customers = Booking_counts[Booking_counts > 1].index

# Filtering dataset to include only these customers
Multiple_booking_data = booking[booking['customer_id'].isin(Multiple_bookings_customers)]

# Count occurrences of each route (from_city to to_city)
Route_counts = Multiple_booking_data.groupby(['from_city', 'to_city']).size().reset_index(name='booking_count')

# Find the most booked route
Most_booked_route = Route_counts.loc[Route_counts['booking_count'].idxmax()]

print(f"Most booked route for customers with more than one booking: {Most_booked_route['from_city']} to {Most_booked_route['to_city']}")
```

## 6) Which are the top 3 departure cities from where customers book mostly in advance, provided that there have been atleast 5 departures from that city?

```python
# Filter bookings where days_to_departure > 0
Advance_Bookings = booking[booking['days_to_departure'] > 0]

# Group by 'from_city' and calculate, Total number of departures, Average number of days customers book in advance
City_Advance_Stats = Advance_Bookings.groupby('from_city').agg(Total_Departures=('booking_id', 'count'), Average_Days_to_Departure=('days_to_departure', 'mean')).reset_index()

# Keep only cities with at least 5 departures
City_Advance_Stats_Filtered = City_Advance_Stats[City_Advance_Stats['Total_Departures'] >= 5]

# Sorting by 'Average_Days_to_Departure' in descending order
City_Advance_Stats_Sorted = City_Advance_Stats_Filtered.sort_values(by='Average_Days_to_Departure', ascending=False)

# Select the top 3 cities
Top_3_cities = City_Advance_Stats_Sorted.head(3)

# Reset index and start numbering from 1
Top_3_cities = Top_3_cities.reset_index(drop=True)
Top_3_cities.index = Top_3_cities.index + 1

# Print the result
print("[Top 3 departure cities where customers book mostly in advance (at least 5 departures)")
print(Top_3_cities)
```

## 7) Plot a heatmap displaying correlations of the numerical column and report which pair of numerical columns in the bookings dataset,have the maximum correlation?

```python
# Selecting only numerical columns from the dataset
numerical_df = booking.select_dtypes(include=[np.number])

# Compute the correlation matrix
correlation_matrix = numerical_df.corr()

# Plot the heatmap
plt.figure(figsize=(12, 8))
# Set a light, clean background
sns.set_style("whitegrid")  
sns.heatmap(correlation_matrix, 
            annot=True,          # Show correlation values
            cmap="coolwarm_r",   # Pastel-reversed coolwarm color palette
            fmt=".2f",           # Format values to 2 decimal places
            linewidths=1,        # Add grid lines for clarity
            square=True,         # Make heatmap cells proportional
            cbar_kws={"shrink": 0.8},  # Adjust color bar size
            annot_kws={"size": 10, "weight": "bold"})  # Make text readable

# Customize the title and labels
plt.title("Correlation Matrix of Numerical Columns in Bookings Data", fontsize=16, fontweight="bold", color="#4a4a4a")
plt.show()

# Find the pair of numerical columns with the highest correlation (excluding self-correlation)
correlation_matrix = correlation_matrix.abs()  # Take absolute values for positive correlation
np.fill_diagonal(correlation_matrix.values, 0)  # Remove diagonal self-correlation

# Get the column pair with maximum correlation
max_corr_pair = correlation_matrix.stack().idxmax()
max_corr_value = correlation_matrix.stack().max()

# Print the result
print(f"The highest correlation is between '{max_corr_pair[0]}' and '{max_corr_pair[1]}' with a correlation of {max_corr_value:.2f}")
```

## 9) Plot the trends at a quarterly frequency for the number of bookings by each of the device types, that is, plot a time series for each year and quarter showing the number of bookings performed by each devicetype.

```python
# Convert 'Booking_Date' to datetime and remove timezone if present
booking['Booking_Date'] = pd.to_datetime(booking['Booking_Date']).dt.tz_localize(None)

# Extract year and quarter from 'Booking_Date'
booking['Year_Quarter'] = booking['Booking_Date'].dt.to_period('Q')

# Group by 'Year_Quarter' and 'Booking_Device' and count bookings
quarterly_trends = booking.groupby(['Year_Quarter', 'Booking_Device']).size().reset_index(name='booking_count')

# Convert 'Year_Quarter' to string for plotting
quarterly_trends['Year_Quarter'] = quarterly_trends['Year_Quarter'].astype(str)

# Plot the time series trends using seaborn
plt.figure(figsize=(12, 6))
sns.lineplot(data=quarterly_trends, x='Year_Quarter', y='booking_count', hue='Booking_Device', marker='o')

# Customize the plot
plt.title("Quarterly Trends of Bookings by Device Type", fontsize=14, fontweight="bold")
plt.xlabel("Year & Quarter", fontsize=12)
plt.ylabel("Number of Bookings", fontsize=12)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.legend(title="Booking_Device")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

## 10) Consider the following example: 12 customers performed a total of 100 searches but only a few of them performed 10 bookings in total. 
## Searches= 100bookings =10customers= 12
## The overall booking to search ratio,Ober,is10/100=0.1 Using the above information/context,answer the following:
## •What is the average oBSR for each month of the year?
## •What is the average oBSR for each day of the week?
## •Plot a time series of oBSR on all the given dates.

```python
# Extract Year-Month and Day of the Week
session['Month'] = session['search_time'].dt.month_name()   # Extracts full month name
session['Day_of_Week'] = session['search_time'].dt.day_name()  # Extracts day name (Monday, Tuesday, etc.)
session['Date'] = session['search_time'].dt.date  # Extract only the date (YYYY-MM-DD)

# Compute total searches and bookings per month
monthly_stats = session.groupby('Month').agg(Total_Searches=('search_id', 'count'), Total_Bookings=('booking_id', 'count')).reset_index()

# Compute Monthly oBSR
monthly_stats['oBSR'] = monthly_stats['Total_Bookings'] / monthly_stats['Total_Searches']

# Compute total searches and bookings per day of the week
daily_stats = session.groupby('Day_of_Week').agg(Total_Searches=('search_id', 'count'), Total_Bookings=('booking_id', 'count')).reset_index()

# Compute Daily oBSR
daily_stats['oBSR'] = daily_stats['Total_Bookings'] / daily_stats['Total_Searches']

# Compute oBSR for each Date (Time Series Analysis)
date_stats = session.groupby('Date').agg(Total_Searches=('search_id', 'count'), Total_Bookings=('booking_id', 'count')).reset_index()

# Compute Time Series oBSR
date_stats['oBSR'] = date_stats['Total_Bookings'] / date_stats['Total_Searches']

# Print results
print("1.Average oBSR for each month:")
print(monthly_stats[['Month', 'oBSR']])

print("\n2.Average oBSR for each day of the week:")
print(daily_stats[['Day_of_Week', 'oBSR']])

# Plot Time Series of oBSR
plt.figure(figsize=(12, 6))
sns.lineplot(data=date_stats, x='Date', y='oBSR', marker='o', color='b')

# Customize the plot
plt.title("3.Time Series of Overall Booking to Search Ratio (oBSR)", fontsize=14, fontweight="bold")
plt.xlabel("Date", fontsize=12)
plt.ylabel("oBSR (Booking/Search Ratio)", fontsize=12)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```
